```python
import pandas as pd
import numpy as np
import cv2
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('hand_gestures.csv')
df.head()
```

```python
df['label'].value_counts()
```

```python
df['label'] = df['label'].map({'Forward': 0, 'Backward': 1, 'Up': 2, 'Down': 3, 'Left': 4, 'Right': 5})
```

```python
input = tf.constant(df)
tensor_data = tf.random.shuffle(input)
```

```python
x_train = tensor_data[:int(len(tensor_data)*0.9), 1:]
y_train = tensor_data[:int(len(tensor_data)*0.9), :1]

x_test = tensor_data[int(len(tensor_data)*0.9):, 1:]
y_test = tensor_data[int(len(tensor_data)*0.9):, :1]
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
a = pd.DataFrame(x_train)
b = pd.DataFrame(x_test)
pd.concat([a,b])
```

```python
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(8, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train, y_train, epochs=100, callbacks=keras.callbacks.EarlyStopping(monitor='accuracy', patience=5))
```

```python
model.evaluate(x_test, y_test)
```

```python
model.save('landmark-model.keras')
```
